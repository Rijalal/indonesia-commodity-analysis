# Indonesian Mining & Commodity Price Analysis
## A Data Science Study of Coal and Nickel Markets (1993–2026)

**Author:** M. Rijal Alghafiqih  
**Dataset:** FRED API World Bank commodity prices + Bank Indonesia exchange rates  
**Tools:** Python · Pandas · Statsmodels · Scikit-learn · Matplotlib · Seaborn  
**Period covered:** January 1993 to May 2026 (401 months)

## Why I Built This Project

Indonesia is the world's largest thermal coal exporter and one of 
the top nickel producers globally. These two commodities are the 
backbone of Indonesia's export economy, yet most data analysis 
I've seen focuses on Western or Chinese markets. I wanted to 
build an analysis specifically through the lens of what these 
price movements mean for Indonesian producers, policymakers, 
and investors.

This project covers 33 years of monthly price data for both 
coal and nickel, combined with Bank Indonesia's historical 
USD/IDR exchange rates. The goal was to go beyond just 
describing what happened and actually build a working 
forecasting model that could be useful for real decisions.

## What I Found 5 Key Insights

### Insight 1: Coal prices are driven by 3 forces, not just supply and demand

When I decomposed the coal price signal I discovered that what 
looks like one price movement is actually three separate forces 
working simultaneously. The long term trend, a consistent 
yearly seasonal cycle, and random shock events that no model 
can predict.

The trend tells me coal has been on a long upward journey 
since 1993, from around $40 per ton to $120-140 today 
even after the post-COVID correction. The seasonal cycle 
repeats every single year with prices typically peaking 
in Q3 before Northern Hemisphere winter. And the shocks, 
GFC in 2008, COVID in 2020, Russia-Ukraine in 2022, 
these are the wild cards that can move prices by 200-300% 
in ways no model can anticipate.

The business implication here is that Indonesian coal 
exporters need two separate strategies. One for riding 
the predictable seasonal cycle, and a completely different 
risk management strategy for surviving the shock events.

### Insight 2: Nickel is fundamentally more volatile than coal

Across all 33 years of data nickel consistently showed 
higher volatility than coal. During the 2008 GFC nickel 
dropped nearly $10,000 in a single month. This isn't 
just statistical noise. It reflects the fundamental 
nature of nickel as a metal with more concentrated 
production, more speculative trading activity, and 
more sensitivity to stainless steel and EV battery 
demand cycles.

For Indonesia this matters enormously right now because 
the country is betting heavily on nickel as the foundation 
of its downstream EV battery industry. Higher volatility 
means higher revenue uncertainty. Indonesian nickel 
producers need stronger hedging strategies and more 
conservative financial planning than their coal counterparts.

### Insight 3: The Rupiah amplifies commodity revenue for Indonesian exporters

One of the most interesting findings in this project was 
how the USD/IDR exchange rate interacts with commodity 
prices. Because commodities are priced in USD but most 
production costs in Indonesia are in Rupiah, a weaker 
Rupiah automatically improves profit margins for exporters 
even when USD prices are flat.

During the post-COVID period I could see this amplification 
effect clearly. Coal prices in IDR terms rose even faster 
than the already dramatic USD price increase. This means 
Indonesian mining companies are naturally hedged in one 
direction. They benefit from both rising commodity prices 
and a weakening Rupiah at the same time.

### Insight 4: Coal prices have extremely strong momentum

Both the PACF analysis and the Random Forest feature 
importance confirmed the same thing. Last month's 
coal price explains nearly 90% of what next month's 
price will be. Coal prices don't randomly jump around. 
Once they start moving in a direction they tend to 
keep moving that way.

This is practically useful. For a procurement team 
buying coal it means if prices are rising, lock in 
contracts now before they go higher. For a producer 
it means if prices are falling, don't wait for a 
recovery that might be months away. Manage costs 
immediately.

### Insight 5: Coal is entering a stable consolidation phase

The 12 month forward forecast from the Random Forest 
model projects coal prices staying in the 138-151 dollar
per ton range through May 2027. This makes intuitive 
sense. After the extraordinary 2022 peak above 400 dollar 
and the subsequent correction back to $100-115 in 
early 2025, coal prices have found a new floor and 
are gradually recovering toward a more sustainable 
long term level.

The slight August 2026 peak in the forecast aligns 
perfectly with the seasonal analysis showing Q3 
as the typical annual high point. No dramatic spikes 
or crashes are predicted under normal market conditions, 
though as the residual analysis showed, black swan 
events remain the biggest wildcard.

## Model Performance Summary

| Metric | Value | Interpretation |
|--------|-------|----------------|
| MAE | 6.58 dollar per ton | Average forecast error |
| RMSE | $9.70 per ton | Penalized error for large misses |
| MAPE | 5.2% | Percentage error, excellent for commodity forecasting |

Professional commodity analysts at major banks typically 
target 15-25% MAPE on coal price forecasts. Achieving 
5.2% on a 12 month test period is a strong result, 
though I want to be honest that this performance 
was measured on a relatively stable price period 
from mid-2025 to mid-2026. During extreme events like 
the 2022 spike, no model would achieve this accuracy.

## Limitations and What I Would Do Next

Every honest analysis needs to acknowledge its limitations. 
Here are mine.

First, the forecasting model only uses price history and 
exchange rates as inputs. A more powerful model would 
incorporate China's industrial production data, global 
energy policy changes, Indonesian export volumes, and 
shipping cost indices. These external factors often 
move prices before the price itself starts moving, 
making them valuable leading indicators.

Second, I only modeled coal price forecasting in detail. 
Nickel deserves its own dedicated forecasting model 
given how different its demand drivers are, especially 
with the EV battery industry creating entirely new 
demand patterns that didn't exist before 2018.

Third, this analysis uses global benchmark prices. 
Indonesian coal trades at a discount to the global 
benchmark depending on calorific value and quality. 
A more precise analysis would use Indonesian-specific 
HBA (Harga Batubara Acuan) prices published monthly 
by the Ministry of Energy.

If I were to extend this project I would add those 
external economic indicators, build a separate nickel 
model, incorporate HBA prices, and deploy the model 
as a simple web dashboard that updates automatically 
every month when new FRED data is released.

## Conclusion

This project taught me that Indonesian commodity markets 
are far more complex and interesting than they appear 
on the surface. The interaction between global demand 
cycles, currency movements, seasonal patterns, and 
random shock events creates a rich analytical challenge 
that I genuinely enjoyed working through.

The 5.2% forecast accuracy feels like a good result, 
but what I am more proud of is understanding WHY the 
model works. Because coal prices have strong momentum 
and the best predictor of tomorrow is today. That kind 
of fundamental insight is more valuable than any 
accuracy metric.

For Indonesia, the story these 33 years of data tell 
is one of growing importance in global commodity markets, 
increasing revenue volatility that needs better risk 
management, and a genuine opportunity in nickel as 
the EV revolution reshapes global metal demand.
