```python
# Coal prices from FRED API - Global coal price index
url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=PCOALAUUSDM"

response = requests.get(url)
from io import StringIO
coal_df = pd.read_csv(StringIO(response.text))
coal_df.columns = ['date', 'coal_price_usd']
coal_df['date'] = pd.to_datetime(coal_df['date'])
coal_df = coal_df[coal_df['coal_price_usd'] != '.'].copy()
coal_df['coal_price_usd'] = pd.to_numeric(coal_df['coal_price_usd'])

print(f"Coal data: {len(coal_df)} months")
print(f"Date range: {coal_df['date'].min()} to {coal_df['date'].max()}")
print(coal_df.tail(5))
```

```python
# USD/IDR exchange rate from FRED - Bank Indonesia data
url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=CCUSMA02IDM618N"

response = requests.get(url)
idr_df = pd.read_csv(StringIO(response.text))
idr_df.columns = ['date', 'usd_idr']
idr_df['date'] = pd.to_datetime(idr_df['date'])
idr_df = idr_df[idr_df['usd_idr'] != '.'].copy()
idr_df['usd_idr'] = pd.to_numeric(idr_df['usd_idr'], errors='coerce')
idr_df = idr_df.dropna()

print(f"USD/IDR data: {len(idr_df)} months")
print(f"Date range: {idr_df['date'].min()} to {idr_df['date'].max()}")
print(idr_df.tail(5))
```

```python
# USD/IDR exchange rate from FRED - Bank Indonesia data
url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=CCUSMA02IDM618N"

response = requests.get(url)
idr_df = pd.read_csv(StringIO(response.text))
idr_df.columns = ['date', 'usd_idr']
idr_df['date'] = pd.to_datetime(idr_df['date'])
idr_df = idr_df[idr_df['usd_idr'] != '.'].copy()
idr_df['usd_idr'] = pd.to_numeric(idr_df['usd_idr'], errors='coerce')
idr_df = idr_df.dropna()

print(f"USD/IDR data: {len(idr_df)} months")
print(f"Date range: {idr_df['date'].min()} to {idr_df['date'].max()}")
print(idr_df.tail(5))
```

```python
# Merge coal + nickel + exchange rate
df = coal_df.merge(nickel_df, on='date', how='inner')
df = df.merge(idr_df, on='date', how='left')

# Forward fill any missing IDR rates
df['usd_idr'] = df['usd_idr'].ffill()

# Convert prices to IDR (in millions for readability)
df['coal_price_idr'] = df['coal_price_usd'] * df['usd_idr']
df['nickel_price_idr'] = df['nickel_price_usd'] * df['usd_idr']

# Add time features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%b')

print(f"Master dataset: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"\nSample:")
print(df[['date', 'coal_price_usd', 'nickel_price_usd', 'usd_idr', 'coal_price_idr']].tail(5))
```

```python
df.to_csv('../data/commodity_master.csv', index=False)
print("Saved!")
print(f"Columns: {df.columns.tolist()}")
print(f"Total rows: {len(df)}")
```
